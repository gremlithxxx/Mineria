```python
# 1. Instalar dependencias
!pip install kagglehub

import pandas as pd
import numpy as np
import kagglehub
from kagglehub import KaggleDatasetAdapter

print("="*50)
print("PARTE 1: IMPORTACIÓN Y EXPLORACIÓN")
print("="*50)

# A. Cargar el dataset directamente desde Kaggle usando kagglehub
print("Cargando dataset desde Kaggle...")
file_path = "student-mat.csv"  # Nombre del archivo específico

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "uciml/student-alcohol-consumption",
  file_path,
)

print("Primeras 5 filas del dataset cargado:")
print(df.head())
print("\n" + "-"*50)

# B. Crear columna student_id usando el índice original
df['student_id'] = df.index
print("Columna 'student_id' creada correctamente")

# C. Revisar dimensiones
print(f"\nDimensiones del DataFrame: {df.shape[0]} filas, {df.shape[1]} columnas")

# D. Identificar tipo de datos de cada columna
print("\nTipos de datos por columna:")
print(df.dtypes)

# E. Mostrar primeras 10 filas
print("\nPrimeras 10 filas del DataFrame:")
print(df.head(10))

# F. Obtener estadísticas descriptivas de columnas numéricas
print("\nEstadísticas descriptivas de columnas numéricas:")
print(df.describe())

# 2. LIMPIEZA DE DATOS

print("\n" + "="*50)
print("PARTE 2: LIMPIEZA DE DATOS")
print("="*50)

# A. Seleccionar aleatoriamente el 5% de posiciones en columnas numéricas y reemplazar por NaN

# Identificar columnas numéricas
columnas_numericas = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Columnas numéricas: {columnas_numericas}")
print(f"Total de columnas numéricas: {len(columnas_numericas)}")

# Crear una copia del DataFrame para no modificar el original
df_limpio = df.copy()

# Calcular el 5% del total de celdas numéricas
total_celdas_numericas = df_limpio[columnas_numericas].size
num_nulos = int(total_celdas_numericas * 0.05)
print(f"\nTotal de celdas numéricas: {total_celdas_numericas}")
print(f"Se insertarán {num_nulos} valores nulos (5%)")

# Generar posiciones aleatorias para insertar NaN
np.random.seed(42)  # Para reproducibilidad
for _ in range(num_nulos):
    fila_aleatoria = np.random.randint(0, df_limpio.shape[0])
    columna_aleatoria = np.random.choice(columnas_numericas)
    df_limpio.loc[fila_aleatoria, columna_aleatoria] = np.nan

print("✓ Valores nulos insertados correctamente")

# B. Contar cantidad de valores nulos

# a. Por columna
print("\na. Valores nulos por columna:")
nulos_por_columna = df_limpio[columnas_numericas].isnull().sum()
nulos_por_columna = nulos_por_columna[nulos_por_columna > 0]
if len(nulos_por_columna) > 0:
    print(nulos_por_columna)
    print(f"Total de columnas con nulos: {len(nulos_por_columna)}")
else:
    print("No se insertaron nulos en ninguna columna (verificar el proceso)")

# b. Por fila
print("\nb. Valores nulos por fila:")
nulos_por_fila = df_limpio[columnas_numericas].isnull().sum(axis=1)
print(f"Filas con al menos un nulo: {(nulos_por_fila > 0).sum()}")
print(f"Porcentaje de filas con nulos: {((nulos_por_fila > 0).sum() / len(df_limpio)) * 100:.2f}%")

# Mostrar distribución de nulos por fila
print("\nDistribución de cantidad de nulos por fila:")
distribucion_nulos = nulos_por_fila.value_counts().sort_index()
for num_nulos, cantidad in distribucion_nulos.items():
    print(f"  {cantidad} fila(s) tienen {num_nulos} valor(es) nulo(s)")

# Información adicional
total_nulos = df_limpio[columnas_numericas].isnull().sum().sum()
print(f"\n RESUMEN FINAL:")
print(f"  • Total de valores nulos insertados: {total_nulos}")
print(f"  • Porcentaje de nulos del total de celdas numéricas: {(total_nulos/total_celdas_numericas)*100:.2f}%")

# Verificar que se insertaron aproximadamente el 5%
print(f"  • Diferencia con el 5% objetivo: {abs((total_nulos/total_celdas_numericas)*100 - 5):.2f}%")

print("\n" + "="*50)
print("FIN DE LA PARTE 1")
print("="*50)

# Mostrar información del DataFrame final
print("\nInformación del DataFrame después de la limpieza:")
print(f"Total de filas: {len(df_limpio)}")
print(f"Total de columnas: {len(df_limpio.columns)}")
print(f"Columnas disponibles: {df_limpio.columns.tolist()}")
```

```python
# PARTE 2: LIMPIEZA DE DATOS (VERSIÓN CORREGIDA)
# Ejecutar después de la Parte 1

print("\n" + "="*50)
print("PARTE 2: LIMPIEZA DE DATOS")
print("="*50)

# A. Seleccionar aleatoriamente el 5% de posiciones en columnas numéricas y reemplazar por NaN

# Identificar columnas numéricas
columnas_numericas = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Columnas numéricas encontradas: {columnas_numericas}")
print(f"Total de columnas numéricas: {len(columnas_numericas)}")

# Crear una copia del DataFrame para trabajar
df_limpio = df.copy()

# Calcular el 5% del total de celdas numéricas
total_celdas_numericas = df_limpio[columnas_numericas].size
num_nulos = int(total_celdas_numericas * 0.05)
print(f"\nTotal de celdas numéricas: {total_celdas_numericas}")
print(f"Se insertarán {num_nulos} valores nulos (5%)")

# Generar posiciones aleatorias para insertar NaN
np.random.seed(42)  # Para reproducibilidad

# Lista para llevar registro de las posiciones modificadas
posiciones_modificadas = []

for i in range(num_nulos):
    fila_aleatoria = np.random.randint(0, df_limpio.shape[0])
    columna_aleatoria = np.random.choice(columnas_numericas)

    # Guardar el valor original antes de modificarlo
    valor_original = df_limpio.loc[fila_aleatoria, columna_aleatoria]

    # Insertar NaN
    df_limpio.loc[fila_aleatoria, columna_aleatoria] = np.nan

    posiciones_modificadas.append({
        'fila': fila_aleatoria,
        'columna': columna_aleatoria,
        'valor_original': valor_original
    })

print("✓ Valores nulos insertados correctamente")

# Mostrar algunas de las posiciones modificadas
print(f"\nPrimeras 10 posiciones modificadas (de {len(posiciones_modificadas)}):")
for i, pos in enumerate(posiciones_modificadas[:10]):
    print(f"  {i+1}. Fila {pos['fila']}, Columna '{pos['columna']}' (valor original: {pos['valor_original']})")

# B. Contar cantidad de valores nulos

# a. Por columna
print("\n" + "-"*30)
print("B.a. VALORES NULOS POR COLUMNA:")
nulos_por_columna = df_limpio[columnas_numericas].isnull().sum()

# Mostrar solo columnas con nulos
nulos_cols_con_valor = nulos_por_columna[nulos_por_columna > 0]
if len(nulos_cols_con_valor) > 0:
    print("\nColumnas numéricas con valores nulos:")
    for columna, cantidad in nulos_cols_con_valor.items():
        porcentaje = (cantidad / len(df_limpio)) * 100
        print(f"  • {columna}: {cantidad} nulos ({porcentaje:.2f}%)")

    print(f"\nTotal de columnas numéricas con nulos: {len(nulos_cols_con_valor)}")
    print(f"Total de columnas sin nulos: {len(columnas_numericas) - len(nulos_cols_con_valor)}")
else:
    print("  No se insertaron nulos en ninguna columna numérica (verificar el proceso)")

# Verificar si hay nulos en columnas no numéricas
nulos_no_numericas = df_limpio.select_dtypes(exclude=[np.number]).isnull().sum()
nulos_no_numericas = nulos_no_numericas[nulos_no_numericas > 0]
if len(nulos_no_numericas) > 0:
    print("\nColumnas NO numéricas con valores nulos (no debería haber):")
    for columna, cantidad in nulos_no_numericas.items():
        print(f"  • {columna}: {cantidad} nulos")

# b. Por fila
print("\n" + "-"*30)
print("B.b. VALORES NULOS POR FILA:")

# Contar nulos por fila (solo en columnas numéricas)
nulos_por_fila = df_limpio[columnas_numericas].isnull().sum(axis=1)

# Estadísticas de nulos por fila
filas_con_nulos = nulos_por_fila[nulos_por_fila > 0]
print(f"Total de filas con al menos un nulo: {len(filas_con_nulos)}")
print(f"Porcentaje de filas con nulos: {(len(filas_con_nulos) / len(df_limpio)) * 100:.2f}%")
print(f"Total de filas sin nulos: {len(df_limpio) - len(filas_con_nulos)}")

# Distribución de cantidad de nulos por fila
print("\nDistribución de cantidad de nulos por fila:")
distribucion_nulos = nulos_por_fila.value_counts().sort_index()
for num_nulos, cantidad in distribucion_nulos.items():
    if num_nulos > 0:
        porcentaje = (cantidad / len(df_limpio)) * 100
        print(f"  • {cantidad} fila(s) tienen {num_nulos} valor(es) nulo(s) ({porcentaje:.2f}%)")
    else:
        print(f"  • {cantidad} fila(s) no tienen nulos ({(cantidad/len(df_limpio))*100:.2f}%)")

# Mostrar las filas con más nulos (VERSIÓN CORREGIDA)
print("\nFilas con mayor cantidad de nulos:")
top_filas_nulos = nulos_por_fila.nlargest(5)
for fila, num_nulos in top_filas_nulos.items():
    if num_nulos > 0:
        print(f"  • Fila {fila}: {num_nulos} nulos")
        # Mostrar qué columnas tienen nulos en esa fila (CORREGIDO)
        columnas_nulas = df_limpio.loc[fila, columnas_numericas].isnull()
        # Obtener los nombres de las columnas donde hay nulos
        columnas_con_nulos = [col for col, es_nulo in zip(columnas_numericas, columnas_nulas) if es_nulo]
        print(f"    Columnas: {', '.join(columnas_con_nulos)}")

# RESUMEN FINAL
print("\n" + "="*30)
print("RESUMEN FINAL DE LIMPIEZA")
print("="*30)

total_nulos = df_limpio[columnas_numericas].isnull().sum().sum()
porcentaje_real = (total_nulos / total_celdas_numericas) * 100

print(f"Total de valores nulos insertados: {total_nulos}")
print(f"Porcentaje de nulos del total de celdas numéricas: {porcentaje_real:.2f}%")
print(f"Porcentaje objetivo: 5%")
print(f"Diferencia: {abs(porcentaje_real - 5):.2f}%")

if abs(porcentaje_real - 5) < 0.1:
    print("✓ El proceso fue exitoso: se insertó aproximadamente el 5% de nulos")
else:
    print("⚠ Hay una diferencia significativa con el objetivo del 5%")

# Mostrar información del DataFrame resultante
print("\n" + "-"*30)
print("INFORMACIÓN DEL DATAFRAME DESPUÉS DE LA LIMPIEZA:")
print(f"Total de filas: {len(df_limpio)}")
print(f"Total de columnas: {len(df_limpio.columns)}")
print(f"Columnas numéricas: {len(columnas_numericas)}")
print(f"Columnas categóricas: {len(df_limpio.columns) - len(columnas_numericas)}")

# Mostrar las primeras filas para verificar
print("\nPrimeras 5 filas del DataFrame limpio:")
print(df_limpio.head())

# Mostrar información de memoria y tipos
print("\nInformación del DataFrame:")
print(df_limpio.info())

# Guardar el DataFrame limpio para usarlo en partes posteriores
print("\n✓ DataFrame 'df_limpio' listo para continuar con las siguientes partes")
```

```python
# PARTE 3: FILTRADO Y SELECCIÓN
# Ejecutar después de haber cargado y limpiado los datos (Partes 1-2)

print("\n" + "="*50)
print("PARTE 3: FILTRADO Y SELECCIÓN")
print("="*50)

# A. Crear columna avg_grade
df_limpio['avg_grade'] = (df_limpio['G1'] + df_limpio['G2'] + df_limpio['G3']) / 3
print("✓ Columna 'avg_grade' creada")
print(f"\nPrimeros 5 promedios:")
print(df_limpio[['G1', 'G2', 'G3', 'avg_grade']].head())

# B. Seleccionar estudiantes con promedio mayor a 12
estudiantes_avg_gt_12 = df_limpio[df_limpio['avg_grade'] > 12]
print(f"\nB. ESTUDIANTES CON avg_grade > 12:")
print(f"   • Cantidad: {len(estudiantes_avg_gt_12)} estudiantes")
print(f"   • Porcentaje del total: {(len(estudiantes_avg_gt_12)/len(df_limpio))*100:.2f}%")
print(f"   • Promedio mínimo en este grupo: {estudiantes_avg_gt_12['avg_grade'].min():.2f}")
print(f"   • Promedio máximo en este grupo: {estudiantes_avg_gt_12['avg_grade'].max():.2f}")
print(f"\n   Primeros 5 estudiantes:")
print(estudiantes_avg_gt_12[['student_id', 'sex', 'avg_grade']].head())

# C. Seleccionar estudiantes con consumo alto de alcohol en fin de semana
estudiantes_walc_alto = df_limpio[df_limpio['Walc'] >= 4]
print(f"\nC. ESTUDIANTES CON Walc >= 4 (consumo alto fin de semana):")
print(f"   • Cantidad: {len(estudiantes_walc_alto)} estudiantes")
print(f"   • Porcentaje del total: {(len(estudiantes_walc_alto)/len(df_limpio))*100:.2f}%")
print(f"   • Valores de Walc en este grupo: {sorted(estudiantes_walc_alto['Walc'].unique())}")
print(f"\n   Primeros 5 estudiantes:")
print(estudiantes_walc_alto[['student_id', 'sex', 'Walc', 'Dalc']].head())

# D. Estudiantes con tiempo de estudio mayor que la mediana
mediana_studytime = df_limpio['studytime'].median()
estudiantes_studytime_gt_med = df_limpio[df_limpio['studytime'] > mediana_studytime]
print(f"\nD. ESTUDIANTES CON studytime > mediana:")
print(f"   • Mediana de studytime: {mediana_studytime}")
print(f"   • Cantidad: {len(estudiantes_studytime_gt_med)} estudiantes")
print(f"   • Porcentaje del total: {(len(estudiantes_studytime_gt_med)/len(df_limpio))*100:.2f}%")
print(f"   • Valores de studytime en este grupo: {sorted(estudiantes_studytime_gt_med['studytime'].unique())}")
print(f"\n   Primeros 5 estudiantes:")
print(estudiantes_studytime_gt_med[['student_id', 'sex', 'studytime', 'avg_grade']].head())

# Información adicional: Comparativa entre grupos
print("\n" + "-"*30)
print("COMPARATIVA RÁPIDA:")
print(f"• Total estudiantes: {len(df_limpio)}")
print(f"• Promedio > 12: {len(estudiantes_avg_gt_12)} ({len(estudiantes_avg_gt_12)/len(df_limpio)*100:.1f}%)")
print(f"• Walc alto: {len(estudiantes_walc_alto)} ({len(estudiantes_walc_alto)/len(df_limpio)*100:.1f}%)")
print(f"• Estudio sobre mediana: {len(estudiantes_studytime_gt_med)} ({len(estudiantes_studytime_gt_med)/len(df_limpio)*100:.1f}%)")
```

```python
# PARTE 4: CREACIÓN DE COLUMNAS NUEVAS
# Ejecutar después de la Parte 3

print("\n" + "="*50)
print("PARTE 4: CREACIÓN DE COLUMNAS NUEVAS")
print("="*50)

# A. Columna high_alcohol (True si Dalc + Walc ≥ 6)
df_limpio['high_alcohol'] = (df_limpio['Dalc'] + df_limpio['Walc']) >= 6
print("A. COLUMNA 'high_alcohol' CREADA")
print(f"   • Total estudiantes: {len(df_limpio)}")
print(f"   • High alcohol (True): {df_limpio['high_alcohol'].sum()} estudiantes")
print(f"   • High alcohol (False): {(~df_limpio['high_alcohol']).sum()} estudiantes")
print(f"   • Porcentaje con high alcohol: {(df_limpio['high_alcohol'].sum()/len(df_limpio))*100:.2f}%")

# Mostrar algunos ejemplos
print(f"\n   Ejemplos de estudiantes con high_alcohol = True:")
ejemplos_high = df_limpio[df_limpio['high_alcohol']].head(3)
for idx, row in ejemplos_high.iterrows():
    print(f"     • Student {row['student_id']}: Dalc={row['Dalc']}, Walc={row['Walc']}, Total={row['Dalc']+row['Walc']}")

# B. Columna alcohol_category con categorías
print(f"\nB. COLUMNA 'alcohol_category' CREADA")

# Definir condiciones
cond_baja = (df_limpio['Dalc'] + df_limpio['Walc']) <= 3
cond_moderada = ((df_limpio['Dalc'] + df_limpio['Walc']) >= 4) & ((df_limpio['Dalc'] + df_limpio['Walc']) <= 5)
cond_alta = (df_limpio['Dalc'] + df_limpio['Walc']) >= 6

# Crear la columna
df_limpio['alcohol_category'] = 'Unknown'  # Valor por defecto
df_limpio.loc[cond_baja, 'alcohol_category'] = 'Low'
df_limpio.loc[cond_moderada, 'alcohol_category'] = 'Moderate'
df_limpio.loc[cond_alta, 'alcohol_category'] = 'High'

# Mostrar distribución
print(f"\n   DISTRIBUCIÓN POR CATEGORÍA:")
categoria_counts = df_limpio['alcohol_category'].value_counts()
for categoria in ['Low', 'Moderate', 'High']:
    count = categoria_counts.get(categoria, 0)
    porcentaje = (count/len(df_limpio))*100
    print(f"   • {categoria}: {count} estudiantes ({porcentaje:.2f}%)")

# Mostrar relación entre Dalc, Walc y categoría
print(f"\n   ESTADÍSTICAS POR CATEGORÍA:")
for categoria in ['Low', 'Moderate', 'High']:
    subset = df_limpio[df_limpio['alcohol_category'] == categoria]
    if len(subset) > 0:
        print(f"   • {categoria}:")
        print(f"     - Dalc promedio: {subset['Dalc'].mean():.2f}")
        print(f"     - Walc promedio: {subset['Walc'].mean():.2f}")
        print(f"     - Total promedio: {(subset['Dalc'] + subset['Walc']).mean():.2f}")

# Mostrar algunos ejemplos
print(f"\n   EJEMPLOS POR CATEGORÍA:")
for categoria in ['Low', 'Moderate', 'High']:
    ejemplos = df_limpio[df_limpio['alcohol_category'] == categoria].head(2)
    print(f"   • {categoria}:")
    for idx, row in ejemplos.iterrows():
        print(f"     - Student {row['student_id']}: Dalc={row['Dalc']}, Walc={row['Walc']}, Total={row['Dalc']+row['Walc']}")
```

```python
# PARTE 5: AGRUPACIONES
# Ejecutar después de la Parte 4

print("\n" + "="*50)
print("PARTE 5: AGRUPACIONES")
print("="*50)

# A. Promedio de avg_grade por sexo
promedio_grade_sexo = df_limpio.groupby('sex')['avg_grade'].mean()
print("A. PROMEDIO DE avg_grade POR SEXO:")
for sexo, promedio in promedio_grade_sexo.items():
    print(f"   • Sexo {sexo}: {promedio:.2f}")

# B. Mediana de Walc por sexo
mediana_walc_sexo = df_limpio.groupby('sex')['Walc'].median()
print(f"\nB. MEDIANA DE Walc POR SEXO:")
for sexo, mediana in mediana_walc_sexo.items():
    print(f"   • Sexo {sexo}: {mediana}")

# C. Número de estudiantes por categoría de alcohol
estudiantes_por_categoria = df_limpio['alcohol_category'].value_counts()
print(f"\nC. NÚMERO DE ESTUDIANTES POR CATEGORÍA DE ALCOHOL:")
for categoria in ['Low', 'Moderate', 'High']:
    count = estudiantes_por_categoria.get(categoria, 0)
    print(f"   • {categoria}: {count} estudiantes")

# D. Promedio de studytime por categoría de alcohol
promedio_studytime_categoria = df_limpio.groupby('alcohol_category')['studytime'].mean()
print(f"\nD. PROMEDIO DE studytime POR CATEGORÍA DE ALCOHOL:")
for categoria in ['Low', 'Moderate', 'High']:
    promedio = promedio_studytime_categoria.get(categoria, 0)
    print(f"   • {categoria}: {promedio:.2f}")

# Información adicional: Tabla resumen completa
print(f"\n" + "-"*30)
print("TABLA RESUMEN COMPLETA:")
resumen_completo = df_limpio.groupby('sex').agg({
    'avg_grade': ['mean', 'median', 'std'],
    'Walc': ['mean', 'median'],
    'Dalc': ['mean', 'median'],
    'studytime': ['mean']
}).round(2)
print(resumen_completo)
```

```python
# PARTE 6: ORDENAMIENTO Y TOP-N
# Ejecutar después de la Parte 5

print("\n" + "="*50)
print("PARTE 6: ORDENAMIENTO Y TOP-N")
print("="*50)

# A. Ordenar estudiantes por avg_grade de mayor a menor
df_ordenado = df_limpio.sort_values('avg_grade', ascending=False)
print("A. ESTUDIANTES ORDENADOS POR avg_grade (MAYOR A MENOR):")
print(f"   • Mejor promedio: {df_ordenado['avg_grade'].iloc[0]:.2f}")
print(f"   • Peor promedio: {df_ordenado['avg_grade'].iloc[-1]:.2f}")

# Crear top 10 y bottom 10
top10 = df_ordenado.head(10).copy()
bottom10 = df_ordenado.tail(10).copy()

print(f"\nB. TOP 10 MEJORES PROMEDIOS:")
print(f"   {'Student ID':<10} {'Sexo':<5} {'G1':<5} {'G2':<5} {'G3':<5} {'Promedio':<8}")
print(f"   {'-'*45}")
for idx, row in top10.iterrows():
    print(f"   {row['student_id']:<10} {row['sex']:<5} {row['G1']:<5} {row['G2']:<5} {row['G3']:<5} {row['avg_grade']:<8.2f}")

print(f"\nC. BOTTOM 10 PEORES PROMEDIOS:")
print(f"   {'Student ID':<10} {'Sexo':<5} {'G1':<5} {'G2':<5} {'G3':<5} {'Promedio':<8}")
print(f"   {'-'*45}")
for idx, row in bottom10.iterrows():
    print(f"   {row['student_id']:<10} {row['sex']:<5} {row['G1']:<5} {row['G2']:<5} {row['G3']:<5} {row['avg_grade']:<8.2f}")

# Estadísticas de los grupos
print(f"\nD. ESTADÍSTICAS DE GRUPOS:")
print(f"   TOP 10:")
print(f"   • Promedio del grupo: {top10['avg_grade'].mean():.2f}")
print(f"   • Desviación estándar: {top10['avg_grade'].std():.2f}")
print(f"   • Distribución por sexo: {top10['sex'].value_counts().to_dict()}")

print(f"\n   BOTTOM 10:")
print(f"   • Promedio del grupo: {bottom10['avg_grade'].mean():.2f}")
print(f"   • Desviación estándar: {bottom10['avg_grade'].std():.2f}")
print(f"   • Distribución por sexo: {bottom10['sex'].value_counts().to_dict()}")
```

```python
# PARTE 7: INTEGRACIÓN DE RESULTADOS (CONCAT Y MERGE)
# Ejecutar después de la Parte 6

print("\n" + "="*50)
print("PARTE 7: INTEGRACIÓN DE RESULTADOS")
print("="*50)

# A. Concatenación de top10 y bottom10
print("A. CONCATENACIÓN DE DATAFRAMES")

# Agregar columna grupo
top10['grupo'] = 'top'
bottom10['grupo'] = 'bottom'

# Concatenar
ranking20 = pd.concat([top10, bottom10], ignore_index=True)
print(f"   • DataFrame ranking20 creado")
print(f"   • Dimensiones: {ranking20.shape[0]} filas, {ranking20.shape[1]} columnas")
print(f"   • Distribución por grupo:")
print(ranking20['grupo'].value_counts())

# B. Tabla resumen por sexo
print(f"\nB. TABLA RESUMEN POR SEXO")
resumen_sex = df_limpio.groupby('sex').agg({
    'avg_grade': 'mean',
    'Walc': 'mean',
    'Dalc': 'mean'
}).round(2)
resumen_sex.columns = ['promedio_avg_grade', 'promedio_Walc', 'promedio_Dalc']
print(resumen_sex)

# C. Unión (merge) entre ranking20 y resumen_sex
print(f"\nC. MERGE ENTRE RANKING20 Y RESUMEN_SEX")
ranking20_enriquecido = pd.merge(ranking20, resumen_sex, on='sex', how='left')
print(f"   • DataFrame ranking20_enriquecido creado")
print(f"   • Dimensiones: {ranking20_enriquecido.shape[0]} filas, {ranking20_enriquecido.shape[1]} columnas")

# Mostrar las nuevas columnas agregadas
print(f"\n   Nuevas columnas agregadas desde resumen_sex:")
nuevas_cols = ['promedio_avg_grade', 'promedio_Walc', 'promedio_Dalc']
for col in nuevas_cols:
    if col in ranking20_enriquecido.columns:
        print(f"   • {col}")

# D. Dimensiones de los DataFrames
print(f"\nD. DIMENSIONES DE LOS DATAFRAMES:")
print(f"   a. ranking20: {ranking20.shape[0]} filas, {ranking20.shape[1]} columnas")
print(f"   b. ranking20_enriquecido: {ranking20_enriquecido.shape[0]} filas, {ranking20_enriquecido.shape[1]} columnas")

# Mostrar contenido de ranking20_enriquecido
print(f"\nCONTENIDO DE RANKING20_ENRIQUECIDO:")
columnas_mostrar = ['student_id', 'sex', 'avg_grade', 'grupo', 'promedio_avg_grade', 'promedio_Walc']
print(ranking20_enriquecido[columnas_mostrar].head(15))
```

```python
# PARTE 8: CRUCE DE VARIABLES
# Ejecutar después de la Parte 7

print("\n" + "="*50)
print("PARTE 8: CRUCE DE VARIABLES")
print("="*50)

# A. Tabla cruzada entre sex y alcohol_category
print("A. TABLA CRUZADA: sex vs alcohol_category")
cruce_sex_alcohol = pd.crosstab(
    df_limpio['sex'],
    df_limpio['alcohol_category'],
    margins=True,
    margins_name='Total'
)
print("\nNúmero de estudiantes por sexo y categoría de alcohol:")
print(cruce_sex_alcohol)

# Versión con porcentajes
print(f"\nPORCENTAJES POR FILA (% dentro de cada sexo):")
cruce_porcentaje_filas = pd.crosstab(
    df_limpio['sex'],
    df_limpio['alcohol_category'],
    normalize='index',
    margins=True,
    margins_name='Total'
).round(4) * 100
print(cruce_porcentaje_filas)

print(f"\nPORCENTAJES POR COLUMNA (% dentro de cada categoría):")
cruce_porcentaje_cols = pd.crosstab(
    df_limpio['sex'],
    df_limpio['alcohol_category'],
    normalize='columns',
    margins=True,
    margins_name='Total'
).round(4) * 100
print(cruce_porcentaje_cols)

# B. Tabla cruzada entre alcohol_category y studytime
print(f"\nB. TABLA CRUZADA: alcohol_category vs studytime")
cruce_alcohol_studytime = pd.crosstab(
    df_limpio['alcohol_category'],
    df_limpio['studytime'],
    margins=True,
    margins_name='Total'
)
print("\nNúmero de estudiantes por categoría de alcohol y tiempo de estudio:")
print(cruce_alcohol_studytime)

# Versión con porcentajes
print(f"\nPORCENTAJES POR FILA (% dentro de cada categoría de alcohol):")
cruce_alcohol_studytime_pct = pd.crosstab(
    df_limpio['alcohol_category'],
    df_limpio['studytime'],
    normalize='index',
    margins=True,
    margins_name='Total'
).round(4) * 100
print(cruce_alcohol_studytime_pct)

# Análisis adicional
print(f"\nANÁLISIS ADICIONAL:")
print("Distribución de studytime por categoría de alcohol:")
for categoria in ['Low', 'Moderate', 'High']:
    subset = df_limpio[df_limpio['alcohol_category'] == categoria]
    if len(subset) > 0:
        print(f"\n  {categoria}:")
        print(f"    • studytime promedio: {subset['studytime'].mean():.2f}")
        print(f"    • studytime mediana: {subset['studytime'].median()}")
        print(f"    • Distribución: {subset['studytime'].value_counts().sort_index().to_dict()}")
```
